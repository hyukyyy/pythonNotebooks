```python
import numpy as np
import pandas as pd
import string
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
fake=pd.read_csv('datasets/Fake.csv',index_col=0)
true=pd.read_csv('datasets/True.csv',index_col=0)
```

```python
news_data=pd.concat([fake,true])
```

```python
news_data.sample(10)
```

```python
news_data[news_data['text']==' ']=np.nan
```

```python
news_data=news_data.dropna()
```

```python
news_data.reindex()
```

```python
porter=PorterStemmer()
stopword=nltk.corpus.stopwords.words('english')
```

```python
def preprocessing(text):
    text=[porter.stem(word) for word in text.lower().split()]
    text=[word for word in text if word not in stopword]
    return text
```

```python
news_data['text']=news_data['text'].apply(preprocessing)
```

```python
news_data.sample(10)
```

```python
from collections import Counter
```

```python
counts=Counter()
```

```python
for ls in news_data['text']:
    counts.update(ls)
```

```python
word_counts=sorted(counts,key=counts.get,reverse=True)
```

```python
word_counts
```

```python
word_to_int={word:ii for ii,word in enumerate(word_counts,1)}
```

```python
word_to_int
```

```python
for ls in news_data['text']:
    for i,word in enumerate(ls):
        ls[i]=word_to_int[word]
```

```python
lensum=0
for ls in news_data['text']:
    lensum+=len(ls)

lensum/len(news_data['text'])
```

```python
sequence_length=250
```

```python
sequences=np.zeros((len(news_data['text']),sequence_length),dtype=int)
```

```python
for i,row in enumerate(news_data['text']):
    text_arr=np.array(row)
    sequences[i,-len(row):]=text_arr[-sequence_length:]
```

```python
from tensorflow.keras import models, layers
```

```python
news_classifier=models.Sequential()
```

```python
n_words=len(word_to_int)+1
```

```python
news_classifier.add(layers.Embedding(n_words,200,embeddings_regularizer='l2'))
```

```python
news_classifier.add(layers.LSTM(16))
```

```python
news_classifier.add(layers.Flatten())
```

```python
news_classifier.add(layers.Dense(1,activation='sigmoid'))
```

```python
news_classifier.summary()
```

```python
news_classifier.compile(loss='binary_crossentropy',optimizer='adam',metrics=['acc'])
```

```python
import time
from tensorflow.keras.callbacks import ModelCheckpoint,TensorBoard
```
